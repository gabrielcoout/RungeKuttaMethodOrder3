## O problema de Modelagem
A modelagem matemática consiste essencialmente em traduzir um problema do mundo real em um problema matemático, resolver o problema matemático e interpretar os resultados.

$$
\begin{array}{ccc}
\fbox{Problema Real} & \xrightarrow{\hspace{2cm}} & \fbox{Modelo Matemático}\\
\Big\uparrow & & \Big\downarrow\\
\fbox{Análise dos Resultados} & \xleftarrow{}  & \fbox{Solução do Modelo}
\end{array}
$$

Um problema do mundo real raramente pode ser traduzido em um problema matemático em toda a sua generalidade. Assim, muitas vezes é necessário "idealizar" ou "simplificar", resultando em um problema que é bastante próximo do original mas que pode ser traduzido e resolvido matematicamente.

## <font color='blue'>O Problema de Transporte de Calor</font>
<img src="cilindrovela.png" alt="Drawing" style="width: 200px;"/>

**Problema**: <br>
Qual a temperatura em cada ponto $x$ da barra cilindrica em cada instante de tempo $t$. 

**Matematicamente**: <br>
Encontrar uma função $u(x,t)$ que informe a temperatura em cada ponto $x$ e tempo $t$.

Simplificações:
1. A barra cilindrica é feita de um material homogêneo
2. A barra cilindrica não gera calor por si só e está isolada do ambiente
3. A barra cilindrica só troca calor com o ambiente em suas extremidades
4. A barra é suficientemente fina para que a temperatura em qualquer ponto de uma seção transversal é constante

**OBS**: As simplificações 3 e 4 tornam o problema unidimensional

Vamos realizar análises quando:
- fixamos $t$ e variamos $x$?
- fixamos $x$ e variamos $t$?

#### Fixando a variável temporal $t$
Quando fixamos $t$, a pergunta natural que aparece é: Qual a temperatura em $x$ neste instante $t$? Não sabemos a temperatura em $x$, mas podemos supor que a temperatura em $x$ é a média da temperatura em uma vizinhança $\Delta x$ de $x$, ou seja:
<img src="cilindrodx.png" alt="Drawing" style="width: 200px;"/>
$$
u(x,t)=\kappa_1\int_{x-\Delta x}^{x+\Delta x}u(x,t)dx
$$

#### Fixando a variável espacial $x$
Quando fixamos $x$, o natural é perguntar o que acontece com a temperatura em $x$ no intervalo de tempo $t+\Delta t$, ou seja, qual a taxa de variação da temperatura em $x$. Podemos assumir que taxa de variação de temperatura em $x$ é proporcional a diferença da "quantidade de temperatura" que flui em $x+\Delta x$ e $x-\Delta x$. 
<img src="flow.png" alt="Drawing" style="width: 200px;"/>
É razoável assumir que "quantidade de temperatura" que flui em um dado local $x$ no instante $t$ é proporcional a $u_x(x,t)$ (derivada com relação a $x$). Assim, temos que:

$$
\frac{u(x,t+\Delta t) - u(x,t)}{\Delta t}\approx\frac{\partial u(x,t)}{\partial t} = \kappa_2(u_x(x-\Delta x)-u_x(x+\Delta x))\\
\Downarrow\\
\kappa_1\frac{\partial}{\partial t}\int_{x-\Delta x}^{x+\Delta x}u(x,t)dx = \kappa_2(u_x(x-\Delta x)-u_x(x+\Delta x))
$$
Passando a derivada temporal para dentro da integral e aplicando o Teorema Fundamental do Cálculo na parte direita da equação:

---
$$
\int_a^b f'(x)dx = f(a)-f(b)
$$

---

$$
\kappa_1\int_{x-\Delta x}^{x+\Delta x}\frac{\partial u(x,t)}{\partial t}dx = \kappa_2\int_{x-\Delta x}^{x+\Delta x} u_{xx}(x,t)dx\\
\Downarrow\\
\int_{x-\Delta x}^{x+\Delta x}\left(\frac{\partial u(x,t)}{\partial t} - \kappa u_{xx}(x,t)\right)dx = 0\\
\Downarrow\\
\frac{\partial u(x,t)}{\partial t} - \kappa u_{xx}(x,t) = 0 \\
\Downarrow \\
u_t(x,t)-\kappa u_{xx}(x,t)=0
$$

$$u_t(x,t)-\kappa u_{xx}(x,t)=0
$$
$$u_t(x,t)=\kappa u_{xx}(x,t)
$$


## <font color='blue'>Soluções analíticas para a Equação - Condições de Contorno</font>
Para solucionar o problema temos que saber quais são as condições no instante $t_0$, o que demanda especificar as condições físicas nesse instante inicial.

Várias possibilidades possíveis:
1. extremidades com temperaturas constantes

$$
u(0,t)=c_0,\quad u(L,t)=c_1  \quad (Dirichilet)
$$

2. extremidades com temperaturas variáveis

$$
u(0,t)=c_0(t),\quad u(L,t)=c_1(t)  \quad (Dirichilet)
$$

$$
u_t(0,t)=c_0(t),\quad u_t(L,t)=c_1(t)  \quad (Newman)
$$

### Extremidades com temperaturas constantes

Suponha $u(0,t)=c_0$ e $u(L,t)=c_1$ (assumimos que a barra tem comprimento $L$). Devido as simplificações do modelo, dado que a temperatura nas extremidades não varia, após algum tempo a distribuição de temperatura alcançará um _estado estacionário_, ou seja, a partir de determinado momento a temperatura não mudará em nenhum ponto da barra. Matematicamente, esta fato significa que:

$$
u_t(x,t) = 0, \quad \forall x\\
\Downarrow\\
u_{xx}(x,t) = u_{xx}(x)=0
$$

Logo, a função $u(x) = a_1x+a_2$ é solução da equação. Sabendo que 
$u(0)=c_0$ e $u(L)=c_1$, temos que:

$$
a_2 = c_0\\
a_1 = \frac{c_1-c_0}{L}
$$

Assim, a solução é:

$$u(x)=\frac{c_1-c_0}{L}x+c_0$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

L = 2
x = np.linspace(0,L,100)
c0 = 3
c1 = 10

u = (c1-c0)/L*x+c0

plt.plot(x,u)
```

### Extremidades com temperaturas variáveis
O caso onde $u(0,t)=c_0(t)$ e $u(L,t)=c_1(t)$ é bem mais difícil de resolver. Por exemplo, assumindo $\kappa=1$, a solução quando $c_0(t)=t$ e $c_1(t)=0$
é dada por:

$$
u(x,t)=(1-\frac{x}{L})t+\frac{2L^2}{\pi^3}\sum_n^\infty\frac{e^{-(\frac{n\pi}{L})^2t}-1}{n^3}sen(\frac{n\pi x}{L})
$$

Veja demonstração [aqui](https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=2&cad=rja&uact=8&ved=2ahUKEwjR3eGqw8PoAhWfDrkGHXRiBEMQFjABegQIARAB&url=https%3A%2F%2Fwww.math.ubc.ca%2F~peirce%2FM257_316_2012_Lecture_20.pdf&usg=AOvVaw39UVxyWrHG29mBa3GtBA8Y)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

numx = 10
numt = 5
L = 2.0
x = np.linspace(0,L,numx)
t = np.arange(0,5)

uat = np.zeros((len(t),len(x)))

for j in range(numt):
    for i in range(numx):
        expres = [((np.exp(-t[j]*(n*np.pi/L)**2)-1.0)/(n**3))*np.sin(n*np.pi*x[i]/L) for n in range(1,10)]
        uat[j,i] += (1.0-x[i]/L)*t[j]+2.0*(L**2)/(np.pi**3)*np.sum(np.asarray(expres))

#print(u)
fig, ax = plt.subplots(1,1)
ax.set_xlabel('x')
ax.set_ylabel('u(x,t)')
for j in range(numt):
    ax.plot(x,uat[j,:],marker='o',label=f't={j}')
    ax.legend()
```

### Diferenças Finitas
Expansão em série de Taylor de uma função $f:\mathbb{R}\rightarrow\mathbb{R}$

$$
f(t+h)=f(t)+hf^\prime(t)+\frac{h^2}{2!}f^{\prime\prime}(t)+\frac{h^3}{3!}f^{\prime\prime\prime}(t)+\cdots
$$

$$
f(t+h) = a_0+a_1h+a_2h^2+\cdots 
$$

$a_0=f(t), \, a_1=f^\prime(t), \, a_2 = \frac{f^{\prime\prime}(t)}{2!}, \ldots$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t =  np.linspace(-3,3,100)
f = np.cos(t)

n=100
h=np.linspace(-2,2,n)
#fh = np.cos(1)*np.ones((n,)) - h*np.sin(1) - h**2/2*np.cos(1) + h**3/6*np.sin(1) + h**4/24*np.cos(1)  # o ponto de aproximacao é 1
fh = np.cos(1)*np.ones((n,)) - h*np.sin(1) - h**2/2*np.cos(1) + h**3/6*np.sin(1) + h**4/24*np.cos(1)  # o ponto de aproximacao é 1

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.scatter(1,np.cos(1),c='r')
ax1.plot(t,f)
ax1.plot(1+h,fh)

ax2.set_xlim(0.5,1.5)
ax2.set_ylim(0.0,1.0)
ax2.scatter(1,np.cos(1),c='r')
ax2.plot(t,f)
ax2.plot(1+h,fh)
```

$$
f(t+h)\approx f(t)+hf^\prime(t)
\newline
\Downarrow
\newline
\frac{f(t+h)-f(t)}{h}\approx f^\prime(t) \\\
\mbox{ aproximação da primeira derivada}
$$

$$
\left\{\begin{array}{c}
f(t+h)\approx f(t)+hf^\prime(t)+\frac{h^2}{2!}f^{\prime\prime}(t) \\
+ \\
f(t-h)\approx f(t)-hf^\prime(t)+\frac{h^2}{2!}f^{\prime\prime}(t) 
\end{array}\right.\\
\Downarrow \\
f(t+h)+f(t-h) \approx 2f(t) + h^2f^{\prime\prime}(t)\\
\Downarrow \\
\frac{f(t+h)-2f(t)+f(t-h)}{h^2}\approx f^{\prime\prime}(t)\\
\mbox{ aproximação da segunda derivada}
$$

### Aproximando a equação do calor por diferenças finitas
$$
u_t(x,t)=\kappa u_{xx}(x,t) 
$$

$$
u_t(x,t)\approx \frac{u(x,t+\Delta t)-u(x,t)}{\Delta t} \\ \\
u_{xx}(x,t)\approx \frac{u(x+\Delta x,t)-2u(x,t)+u(x-\Delta x,t)}{\Delta x^2}
$$

$$
\Downarrow\\
\frac{u(x,t+\Delta t)-u(x,t)}{\Delta t}\approx \kappa \frac{u(x+\Delta x,t)-2u(x,t)+u(x-\Delta x,t)}{\Delta x^2}
$$

Discretizando o intervalo de tempo $[0,T]$ e a barra tem comprimento $[0,L]$ em pontos $0=t_0,t_1,\cdots,t_k=T$ e $0=x_0,x_1,\cdots,x_n=L$,
respectivamente:

$$
u(x_i,t_{i+1}) = u(x_i,t_i) +\Delta t\left(\kappa \frac{u(x_{i+1},t_i)-2u(x_i,t_i)+u(x_{i-1},t_i)}{\Delta x^2}\right)
$$

Se soubermos os valores $u(x_i,t_0)$, $\forall i$ então podemos calcular $u(x_i,t_1),u(x_i,t_2),\ldots,u(x_i,t_T)$ utilizando a fórmula acima.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# discretização da variável temporal (t)
nT = 900
T = 2
dt = T/nT
t = np.linspace(0,T,nT)

# discretização a variável espacial (x)
nx = 30
L = 2
dx = L/nx
x = np.linspace(0,L,nx)

kappa = 1
u = np.zeros((nx,nT))  # cada coluna corresponde a solução em um determinado instante t

# condições de contorno constantes
u[0,:] = 3.0    
u[-1,:] = 10.0

for i in range(1,nT):
    for j in range(1,nx-1):
        u[j,i] = u[j,i-1] + dt*(kappa*(u[j+1,i-1]-2.0*u[j,i-1]+u[j-1,i-1])/dx**2)
u
np.set_printoptions(precision=2)
# # print(u[:50,:6])

for i in range(0,nT,50):
    plt.plot(x,u[:,i])
```

```python
import matplotlib.patches as patches
# calcular o erro de aproximacao

L = 2
xa = np.linspace(0,L,30)
c0 = 3
c1 = 10

ua = (c1-c0)/L*xa+c0

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(6,4))
ax1.plot(xa,ua)
ax1.plot(x,u[:,-1])
rect = patches.Rectangle((0.75,6),0.5,1,linewidth=1,edgecolor='k',facecolor='none')
ax1.add_patch(rect)

ax2.set_xlim(0.75,1.25)
ax2.set_ylim(6,7)
ax2.plot(xa,ua)
ax2.plot(x,u[:,-1])

print(ua.shape,u.shape)
print("erro de aproximacao:",np.linalg.norm(ua-u[:,-1])**2)
```

## <font color='blue'> Consistência, Estabilidade e Convergência do Método Numérico </font>

A equação do calor, como descrita acima, é uma equação diferencial parcial do tipo parabólica (existem outros dois tipos de equações diferencias, as hiperbólicas e as elípticas). 

Quando resolvemos uma equação diferencial parcial numericamente, devemos sempre considerar três aspectos importantes:

1. **Consistência**: um esquema numérico de solução é dito consistente quando a equação discretizada converge para a equação original (contínua) quando $\Delta t$ e $\Delta x$ tendem a zero.

2. **Estabilidade**: um esquema numérico é estável se a diferença entre a solução numérica e a solução exata permanece limitada quando o número de passos do método numérico tende a infinito

3. **Convergência**: um esquema numérico é convergente se a diferença entre a solução numérica e a solução exata tende a zero, em todos os pontos, quando $\Delta t$ e $\Delta x$ tende a zero. Ou seja, se $u(x,t)$ é a solução exata e $\tilde{u}(x,t)$ é a solução aproximada, então:
$$
\lim_{\Delta t,\Delta x\rightarrow 0} \|u-\tilde{u}\|=0
$$

'>>>' Convergência é mais forte

No caso particular do método numérico gerado pela discretização por diferenças finitas

$$
u(x_i,t_{i+1}) = u(x_i,t_i) +\Delta t\left(\kappa \frac{u(x_{i+1},t_i)-2u(x_i,t_i)+u(x_{i-1},t_i)}{\Delta x^2}\right)
$$

pode-se mostrar que ele é consistente e que estabilidade é garantida se 

$$
\kappa\frac{\Delta t}{\Delta x^2} \leq \frac{1}{2}
$$

Um importante resultado no contexto de equações diferenciais parciais é o teorema de Lax, que garante que um método é convergente se e só se ele é estável e consistente. Logo, como o método descrito acima é estável e consistente, temos que ele é convergente.

### Exemplo:
Vamos supor que as condições de contorno são dadas por duas funções

$$
c_0(t)=\sin(t),\quad c_1(t)=t^2, \quad t\in[0,3]
$$

```python
nT = 200
T = 3.0
dt = T/nT
t = np.linspace(0,T,nT)

nx = 20
L = 2
dx = L/nx
x = np.linspace(0,L,nx)

kappa = 1

u = np.zeros((nx,nT))
u[0,:] = np.sin(t)
u[-1,:] = t**2
for i in range(1,nT):
    u[1:-1,i]=u[1:-1,i-1]+dt*kappa*np.convolve(u[:,i-1],[1,-2,1],'valid')/dx**2

np.set_printoptions(precision=2)
# print(u[:50,:6])

for i in range(0,nT,10):
    plt.plot(x,u[:,i])
```

Note que a solução encontrada não está correta. A razão é que a condição para estabilidade

$$
\kappa\frac{\Delta t}{\Delta x^2}\leq\frac{1}{2}
$$

não é satisfeita.

```python
print('Verificando a condição de estabilidade',kappa*dt/dx**2)
```

```python
# nT = 600
T = 3.0
# dt = T/nT
# t = np.linspace(0,T,nT)

nx = 10
L = 2
dx = L/nx
x = np.linspace(0,L,nx)

kappa = 1

dt = dx**2/(2*kappa)
nT = int(T/dt)
t = np.linspace(0,T,nT)


print('Verificando a condição de estabilidade',kappa*dt/dx**2)

u = np.zeros((nx,nT))
u[0,:] = np.sin(t)
# u[0,:] = t**3
u[-1,:] = t**2
for i in range(1,nT):
    u[1:-1,i]=u[1:-1,i-1]+dt*kappa*np.convolve(u[:,i-1],[1,-2,1],'valid')/dx**2

np.set_printoptions(precision=2)
# print(u[:50,:6])

for i in range(0,nT,20):
    plt.plot(x,u[:,i])
```
